```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
# Grafico de franjas / Bandas 
import seaborn as sns
sns.stripplot(data = df, y="writing score", x="math score", jitter=True, palette="husl")
```

```python
# Grafico de franjas / Bandas con matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.stripplot(data = df, y="writing score", x="math score", jitter=True, palette="Set3")
ax.axes.xaxis.set_ticks([])
plt
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.stripplot(data = df, y="writing score", x="race/ethnicity", jitter=True, palette="Set1")
plt.show
```

```python
# Diagrama de enjambre (swarmplot)
# Busca colocar las observaciones  de manera que no se empalmen
sns.swarmplot(data=df, y="writing score", x="race/ethnicity", palette="Set2")
```

```python
print(sns.__version__)
```

```python
# Diagrama de Caja (boxplot)
sns.boxplot(data=df, y="writing score", x="race/ethnicity", hue="gender")
```

```python
#  Grafico de Violin 
# Mezcla de un boxplot con funcion de densidad Kernel donde mas ancho es el segmento existe una mayor concentracion de valores 
sns.violinplot(data=df, y="writing score", x="race/ethnicity", hue="gender")
```

```python
#Letter-Value Plot (lvplot)
# Mezcla de boxplot y el violinplot, conveniente para basesde datos grandes 
sns.catplot(x="race/ethnicity", y="writing score", kind="boxen", data=df, hue="gender")
```

```python
# Grafico de barras (barplot) para observar la mediana de cada valor 
sns.barplot(data=df, y="writing score", x="race/ethnicity", hue="lunch")
```

```python
# Grafico de Puntos
sns.pointplot(data=df, y="writing score", x="race/ethnicity", hue="lunch")
```

```python
#Grafico de conteo de puntos(countplot)
sns.countplot(data=df, y="race/ethnicity", hue="lunch")
```

```python
# Graficos de regrecion (regplot)
sns.regplot(data=df, y="writing score", x="reading score", marker="+")
```

```python
# Grafico de residuales 
sns.residplot(data=df, y="writing score", x="reading score")
```

```python
# Grafico de regresion con Regplot y media por valor de x
# Muy util para enfatizar tendencias
import numpy as np
sns.regplot(data=df, y="writing score", x="reading score", x_estimator=np.mean)
```

```python
# Mapa de Calor (heatmap)
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr())
```

```python
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True)
```

```python
# Tablas Cruzadas
pd.crosstab(df["gender"], df["race/ethnicity"], values=df["math score"], aggfunc="mean")
```

```python
sns.heatmap(pd.crosstab(df["gender"], df["race/ethnicity"], values=df["math score"], aggfunc="mean"), annot=True)
```
