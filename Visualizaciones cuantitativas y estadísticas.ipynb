```python
#Asignacion de indice titulos e indices con index_col
import pandas as pd
df = pd.read_csv("olympic-medals.csv", index_col=0)
df
```

```python
# los 10 primeros valores
medallas = df.head(10)
medallas
```

```python
# Visualizacion de medallas de Oro y Plata
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
plt.show()
```

```python
# Visualizacion de medallas de Oro y Plata con titulos de eje x en vertical
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
# Visualizacion de medallas de Oro y Plata Grafico de barras apiladas con las medallas de oro como base con la funcion bottom
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
# Visualizacion de medallas de Oro y Plata y bronce apiladas notese que es como anidar funciones 
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
# Visualizacion de medallas de Oro y Plata y bronce Adicion de cuado de LEYENDA
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
ax.legend(["Gold","Silver","Bronze"])
plt.show()
```

```python
# Creacion de un histograma 
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"])
plt.show
```

```python
# Creacion de dos histogramas en un solo grafico con leyendas utilizando la funcion label y ax.legend
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro")
ax.hist(df["Silver"], label ="Plata")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show
```

```python
# Definicion de intervalos para los ejes del grafico 

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro", bins = [0,5,10,15,20,25,30,35,40])
ax.hist(df["Silver"], label ="Plata", bins = [0,5,10,15,20,25,30,35,40])
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show
```

```python
# Histograma con transparencias 

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro", bins = [0,5,10,15,20,25,30,35,40], histtype = "step")
ax.hist(df["Silver"], label ="Plata", bins = [0,5,10,15,20,25,30,35,40], histtype = "step")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show
```

```python
# Diagrama de caja
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
ax.boxplot([df["Gold"], df["Silver"]])
ax.set_xticklabels(["Oro","Plata"])
ax.set_ylabel("# de medallas")
plt.show
```

```python
# Grafico de dispersion
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"])
ax.set_ylabel("# de medallas de oro")
ax.set_ylabel("# de medallas de plata")
plt.show
```

```python
# Grafico de dispersion multiple
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"], color = "red", label ="Plata")
ax.scatter(df["Gold"], df["Bronze"], color = "Blue", label ="Bronce")
ax.set_xlabel("# de medallas de oro")
ax.set_ylabel("# de medallas de plata")
ax.legend()
plt.show
```
