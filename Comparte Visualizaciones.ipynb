##### 1. Grfico de lneas bsico

```python
import matplotlib.pyplot as plt

# Datos para el grfico

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

# Crear el grfico

plt.plot(x, y, label="Doble de X")

plt.title("Grfico de lneas")

plt.xlabel("Eje X")

plt.ylabel("Eje Y")

plt.legend()

plt.show()
```

##### 2. Grfico de barras

```python
# Datos

categorias = ["A", "B", "C", "D"]

valores = [3, 7, 5, 9]

# Crear el grfico

plt.bar(categorias, valores, color="skyblue")

plt.title("Grfico de barras")

plt.xlabel("Categoras")

plt.ylabel("Valores")

plt.show()
```

##### 3. Grfico de dispersin (scatter plot)

```python
import numpy as np

# Datos aleatorios

x = np.random.rand(50)

y = np.random.rand(50)

tamao = np.random.rand(50) * 100

# Crear el grfico

plt.scatter(x, y, s=tamao, c="purple", alpha=0.6, edgecolors="black")

plt.title("Grfico de dispersin")

plt.xlabel("Eje X")

plt.ylabel("Eje Y")

plt.show()
```

##### 4. Grfico de pastel

```python
# Datos

etiquetas = ["Manzanas", "Bananas", "Naranjas", "Uvas"]

porcentajes = [30, 25, 20, 25]

colores = ["red", "yellow", "orange", "purple"]

# Crear el grfico

plt.pie(porcentajes, labels=etiquetas, colors=colores, autopct="%1.1f%%", startangle=90)

plt.title("Distribucin de frutas")

plt.axis("equal") # Asegura que el grfico sea un crculo

plt.show()
```

##### 5. Histogramas

```python
# Datos

datos = np.random.normal(loc=50, scale=10, size=1000)

# Crear el histograma

plt.hist(datos, bins=20, color="green", edgecolor="black", alpha=0.7)

plt.title("Histograma de datos")

plt.xlabel("Valores")

plt.ylabel("Frecuencia")

plt.show()
```

##### 6. Grficos combinados

```python
# Datos

x = np.arange(1, 11)

y1 = x ** 2

y2 = x ** 3

# Crear dos grficos en el mismo espacio

plt.plot(x, y1, label="X^2", color="blue")

plt.plot(x, y2, label="X^3", color="red")

plt.title("Grficos combinados")

plt.xlabel("Eje X")

plt.ylabel("Valores")

plt.legend()

plt.show()
```

##### 7. Subgrficos (multiple plots)

```python
# Datos

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

y2 = np.cos(x)

# Crear subgrficos

plt.figure(figsize=(10, 5))

# Subgrfico 1

plt.subplot(1, 2, 1)

plt.plot(x, y1, color="blue")

plt.title("Seno")

# Subgrfico 2

plt.subplot(1, 2, 2)

plt.plot(x, y2, color="red")

plt.title("Coseno")

plt.tight_layout()

plt.show()
```

##### 8. Personalizacin avanzada con estilos

```python
# Aplicar un estilo de Matplotlib

plt.style.use("seaborn-darkgrid")

# Datos

x = np.linspace(0, 10, 100)

y = np.exp(-x) np.sin(2 np.pi * x)

# Crear el grfico

plt.plot(x, y, label="Decaimiento exponencial", linewidth=2, color="teal")

plt.title("Grfico con estilo personalizado")

plt.xlabel("Tiempo")

plt.ylabel("Amplitud")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Datos de ejemplo
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

def update(frame):
    line.set_ydata(np.sin(x + frame / 10))  # Actualiza los datos de la lnea
    return line,

ani = FuncAnimation(fig, update, frames=100, blit=True)
ani.save('animacion.gif', writer='imagemagick')
```
